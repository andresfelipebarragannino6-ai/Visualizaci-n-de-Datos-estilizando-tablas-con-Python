```python
import pandas as pd

# Preparación de fechas
tienda_libros['fecha_llegada'] = pd.to_datetime(tienda_libros['fecha_llegada'])
tienda_libros['fecha_pedido'] = pd.to_datetime(tienda_libros['fecha_pedido'])

fechas = tienda_libros.copy()
fechas = fechas.sort_values('fecha_pedido')
fechas = fechas[fechas['fecha_pedido'] >= '2013-01-01']

fechas['mes'] = fechas['fecha_pedido'].dt.strftime('%Y - %b')
fechas['tiempo_entrega'] = (
    fechas['fecha_llegada'] - fechas['fecha_pedido']
).dt.days

# Tabla dinámica
entregas = fechas.pivot_table(
    index='pais',
    columns='mes',
    values='tiempo_entrega',
    aggfunc='mean',
    sort=False
)

# Definición de estilos
columas = {
    'selector': '.col_heading',
    'props': 'font-weight:normal; font-family:Arial'
}

tabla = {
    'selector': 'td, th',
    'props': 'text-align:left'
}

indices = {
    'selector': '.index_name',
    'props': 'font-weight:bold; font-family:Arial; text-align:right'
}

filas_gris = {
    'selector': (
        'tbody tr:nth-child(2) th.row_heading,'
        'tbody tr:nth-child(4) th.row_heading,'
        'tbody tr:nth-child(6) th.row_heading,'
        'tbody tr:nth-child(8) th.row_heading'
    ),
    'props': 'background-color:#2b2b2b; color:white'
}

# Aplicación de estilos
estilo_mensual = entregas.style.format('{:.0f}')

estilo_mensual = estilo_mensual.set_table_styles(
    [columas, tabla, indices, filas_gris],
    overwrite=False
)

estilo_mensual = estilo_mensual.set_sticky(axis='index')

estilo_mensual
```

```python
import pandas as pd

ventas = pd.DataFrame({
    "segmento_cliente": ["B2B", "B2C", "Retail"],
    "modo_envio": ["Express", "Normal", "Express"],
    "ventas": [1000, 1500, 800]
})

tabla = (
    ventas
    .pivot_table(
        index="segmento_cliente",
        columns="modo_envio",
        values="ventas",
        aggfunc="sum"
    )
    .assign(Total=lambda df: df.sum(axis=1))
)

tabla.loc["Total"] = tabla.sum()

def pintar_nulos(valor):
    return "background-color: #FF6B6B" if pd.isna(valor) else ""

estilo = (
    tabla
    .style
    .format("{:,.2f}")
    .applymap(pintar_nulos)
    .set_table_styles(
        [{
            "selector": "tr:last-child",
            "props": [("background-color", "#4682B4"), ("color", "white")]
        }],
        overwrite=False
    )
)

estilo
```
